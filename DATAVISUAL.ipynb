```
from collections import Counter
import pandas as pd
import seaborn as sn
import numpy as np
import csv
```

```
font = {'family': 'serif',
        'color':  'Red',
        'weight': 'normal',
        'size': 16,
        }
```

```
font1 = {'family': 'serif',
        'color':  'green',
        'weight': 'normal',
        'size': 18,
        }
```

```
font2 = {'family': 'serif',
        'color':  'Black',
        'weight': 'normal',
        'size': 18,
        }
```

```
x = []
y = []

with open('/content/df1.csv', 'r') as csvfile:
  plots = csv.reader(csvfile, delimiter=',')
  for row in plots:
        x.append(row[0])
        y.append(row[1])
```

```
plt.bar(x, y)
plt.xticks(rotation=90)
plt.xlabel('Genre', fontdict=font)
plt.ylabel('Jumlah genre', fontdict=font1)
plt.title('Top genre in Top 10 Music chart in 2010', fontdict=font2)
plt.show()
```
